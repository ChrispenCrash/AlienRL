```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
x = np.arange(-3, 3.1, 0.1)
y = np.tanh(x)
```

```python
combo = ['science']
plt.style.use(combo)

with plt.style.context(combo):
    plt.figure(figsize=(10, 5))
    plt.title('Tanh Function', fontsize=20, fontweight='heavy')
    plt.plot(x, y)
    plt.hlines(y=0, xmin=-3, xmax=3, ls='-', lw=1, color='black')
    plt.vlines(x=0, ymin=-1, ymax=1, ls='-', lw=1, color='black')
    plt.xlabel('x', fontsize=15, fontweight='bold')
    plt.ylabel('tanh(x)', fontsize=15, fontweight='bold', rotation=0, labelpad=20, va='center')
    plt.xlim(-3, 3)
    plt.ylim(-1, 1)
    plt.grid()
    plt.show()
```
